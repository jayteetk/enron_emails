## Creating the full topic matrix.

As we have split the initial 200,000 unique, meaningful emails into 3 parts for application of our LDA model, and exported the topical results into a seperate csv, we are importing them here and merging for batch processing. We would be applying the following 

1) Merge all 3 topic matrices
2) Plot the distribution of topics to identify any possible anomalies. 
3) Apply a threshhold tolerance of mean + 0.25 * std deviation for each topic. (to improve quality of results)

The result of this exercise presents us with about 145k of higher quality topic attributions, which we would use for analysis, anything else is deemed as 'Unclassified'

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/Capstone/dataset/topics1.csv')
df2 = pd.read_csv('D:/Capstone/dataset/topics2.csv')
df3 = pd.read_csv('D:/Capstone/dataset/topics3.csv')
```

```python
print(len(df),len(df2),len(df3))
print('total:',len(df)+len(df2)+len(df3))
```

```python
df = df.append(df2, ignore_index= True)
df = df.append(df3, ignore_index= True)
```

```python
topic_dict={0:'office communications',
                1:'industry expertise',
                2:'social',
                3:'operational compliance',
                4:'trading',
                5:'legal',
                6:'spam & html communique'
               }
```

```python
df= df.drop(columns = 'Unnamed: 0')
```

```python
df.columns = ['topic_0','topic_1','topic_2','topic_3','topic_4','topic_5','topic_6']
```

```python
df['pred_topic'] = df.idxmax(axis=1)
df['max_prob'] = df.max(axis=1)
```

```python
topic_list=['topic_0','topic_1','topic_2','topic_3','topic_4','topic_5','topic_6']

def summary_table(df,cls,pb_cl,topic_list):
    summary=pd.DataFrame({})
    for i in topic_list:
        items=[]
        mean   = df[df[cls]==i][pb_cl].mean()
        items.append(mean)
        sd   = df[df[cls]==i][pb_cl].std()
        items.append(sd)
        qpct   = np.percentile(df[df[cls]==i][pb_cl], 25)
        items.append(qpct)
        median = df[df[cls]==i][pb_cl].median()
        items.append(median)
        tqpct  = np.percentile(df[df[cls]==i][pb_cl], 75)
        items.append(tqpct)
        th     = mean + sd*(.25)
        items.append(th)

        summary[i]=items
    summary.rename({0: 'mean', 1: 'sd', 2:'25%',3:'50%',4:'75%',5:'threshhold'}, axis='index',inplace=True)
    return summary
```

```python
summary=summary_table(df,'pred_topic','max_prob',topic_list)
```

```python
threshhold=summary.T.threshhold.tolist()
```

```python
summary.T['75%'].mean()
```

```python
alpha=summary.T.threshhold.mean()
```

```python
print(round(df.pred_topic.value_counts(normalize=True)*100,3))
df_check=df[df['max_prob']>=alpha]
print(round(df_check['pred_topic'].value_counts(normalize=True)*100,3))
```

```python
def final_topic(probas, topic, threshhold):
    final=[]
    for i in range(len(probas)):
        if topic[i] == 'topic_0' and probas[i]>= threshhold[0]:
            final.append(topic[i])
        elif topic[i] == 'topic_1' and probas[i]>= threshhold[1]:
            final.append(topic[i])
        elif topic[i] == 'topic_2' and probas[i]>= threshhold[2]:
            final.append(topic[i])
        elif topic[i] == 'topic_3' and probas[i]>= threshhold[3]:
            final.append(topic[i])
        elif topic[i] == 'topic_4' and probas[i]>= threshhold[4]:
            final.append(topic[i])
        elif topic[i] == 'topic_5' and probas[i]>= threshhold[5]:
            final.append(topic[i])
        elif topic[i] == 'topic_6' and probas[i]>= threshhold[6]:
            final.append(topic[i])
        else:
            final.append('unclassified')
    return final
```

```python
df['final_topic']= final_topic(df.max_prob,df.pred_topic,threshhold)
```
