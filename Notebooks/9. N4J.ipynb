dfr does not have 'unclassified' topics, we would want to include these only for intra-network analysis.
Lets extract this from df where it includes all topics, including 'unclassified' ones.
Let us also include this into the reduced dataframe. 

```python
dfr_rec=dfr[dfr.type=='received']
df_rec =df[df.type=='received']
```

```python
dfr_rec_a= pd.DataFrame()
df_rec_a= pd.DataFrame()
```

```python
for a in a_list:
    result = dfr_rec[(dfr_rec['eid']==a)]
    dfr_rec_a=dfr_rec_a.append(result)

for a in a_list:
    result = df[(df['eid']==a)]
    df_rec_a=df_rec_a.append(result)
```

```python
print(len(dfr_rec_a))
print(len(df_rec_a))
print(len(dfr_rec_a.eid.unique()))
print(len(df_rec_a.eid.unique()))
```

```python
print(len(dfr_rec_a.eid.unique()))
print(len(dfr_rec_a.uuid.unique()))
```

```python
print(len(df_rec_a.eid.unique()))
print(len(df_rec_a.uuid.unique()))
```

```python
missing_a=[i for i in df_rec_a.eid.unique().tolist() if i not in dfr_rec_a.eid.unique().tolist()]
```

```python
missing_df=pd.DataFrame()
for m in missing_a:
    result = df_rec_a[(df_rec_a['eid']==m)&(df_rec_a['topic']=='unclassified')]
    missing_df=missing_df.append(result)
```

```python
print(len(missing_df))
print(len(missing_df.uuid.unique()))
print(len(missing_df.eid.unique()))
print(len(missing_a))
```

```python
dfr_rec_a = pd.concat([dfr_rec_a,missing_df],axis=0)
print(len(dfr_rec_a.eid.unique()))
print(len(dfr_rec_a.uuid.unique()))
```

```python
missing_u=[i for i in df_rec_a.uuid.unique().tolist() if i not in dfr_rec_a.uuid.unique().tolist()]
```

```python
missing_df=pd.DataFrame()
for m in missing_u:
    result = df_rec_a[(df_rec_a['uuid']==m)&(df_rec_a['topic']=='unclassified')]
    missing_df=missing_df.append(result)
```

```python
dfr_rec_a = pd.concat([dfr_rec_a,missing_df],axis=0)
print(len(dfr_rec_a.eid.unique()))
print(len(dfr_rec_a.uuid.unique()))
```
