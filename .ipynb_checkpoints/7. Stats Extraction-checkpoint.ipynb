```python
def sent_count(df,uuid):
    eids= df[(df['uuid']== uuid)&(df['type']=='sent')]['eid'].unique().tolist()
    count_list=[]
    for i in eids:
        count = len((df[(df['uuid']== uuid)&(df['type']=='sent')&(df['eid']==i)]))
        count_list.append(count)
    result = pd.DataFrame({'eid':eids,str(uuid):count_list})
    return result
```

```python
def sent_count_topic(df,uuid,topic):
    eids= df[(df['uuid']== uuid)&(df['type']=='sent')]['eid'].unique().tolist()
    count_list=[]
    for i in eids:
        count = len((df[(df['uuid']== uuid)&(df['type']=='sent')&(df['eid']==i)&(df['final_topic']==topic)]))
        count_list.append(count)
    result = pd.DataFrame({'eid':eids,str(uuid):count_list})
    return result
```

```python
def rec_count(df,uuid):
    eids= df[(df['uuid']== uuid)&(df['type']=='received')]['eid'].unique().tolist()
    count_list=[]
    for i in eids:
        count = len((df[(df['uuid']== uuid)&(df['type']=='received')&(df['eid']==i)]))
        count_list.append(count)
    result = pd.DataFrame({'eid':eids,str(uuid):count_list})
    return result
```

```python
def rec_count_topic(df,uuid,topic):
    eids= df[(df['uuid']== uuid)&(df['type']=='received')&(df['final_topic']==topic)]['eid'].unique().tolist()
    count_list=[]
    for i in eids:
        count = len((df[(df['uuid']== uuid)&(df['type']=='received')&(df['eid']==i)&(df['final_topic']==topic)]))
        count_list.append(count)
    result = pd.DataFrame({'eid':eids,str(uuid):count_list})
    return result
```
